```python
from langgraph.graph import StateGraph , START , END
#from langchain_openai import ChatOpenAI
from langchain_community.chat_models import ChatOllama
from typing import TypedDict
from dotenv import load_dotenv
```

```python
model = ChatOllama(model="phi")
```

```python
class BlogState(TypedDict):
    title : str
    outline : str
    content : str
    score : int
```

```python
def create_outline(state : BlogState) -> BlogState:
    title = state['title']
    prompt = f'Generate a detailed outline for a blog on the topic - {title} based on your understanding and context you have'
    outline = model.invoke(prompt).content
    state['outline']=outline
    return state
```

```python
def create_blog(state : BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']
    prompt = f'Write a detailed blog on the - {title} based on the following outline -{outline} based on your understanding.'
    content = model.invoke(prompt).content
    state['content']=content
    return state
```

```python
def evaluate_blog(state : BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']
    content = state['content']

    prompt = f'Based on the title - {title}, you have to rate this content \n {content} \n . Rate on a scale of 1-10 while giving appropriate reasons. Return only an integer as your answer'
    score = model.invoke(prompt).content
    state['score']=score
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node('create_outline',create_outline)
graph.add_node('create_blog',create_blog)
graph.add_node('evaluate_blog',evaluate_blog)

graph.add_edge(START,'create_outline')
graph.add_edge('create_outline','create_blog')
graph.add_edge('create_blog','evaluate_blog')
graph.add_edge('evaluate_blog',END)

workflow = graph.compile()
```

```python
initial_state = {'title':"Rise of AI in India"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
